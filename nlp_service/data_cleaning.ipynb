```python
#supprimer les colonnes qui sont presque vide
cols_to_drop = ['value_price_usd', 'sale_price_usd', 'variation_desc', 'child_max_price', 'child_min_price']
df = df.drop(columns=cols_to_drop, errors='ignore')
```

```python
# REVIEWS  : Toujours 0 si vide
df['reviews'] = df['reviews'].fillna(0)
```

```python
global_mean_rating = round(df['rating'].mean(), 2)
def fill_rating(row):
        # Si la note existe déjà, on la garde
        if pd.notnull(row['rating']) and row['rating'] != 0:
            return row['rating']

        # Si la note est manquante (ou 0) :
        if row['new'] == 1:
            return global_mean_rating # C'est nouveau -> On encourage (Note Moyenne)
        else:
            return 0 # C'est vieux et sans avis -> On pénalise (0)
df['rating'] = df.apply(fill_rating, axis=1)
```

```python
df['highlights'].unique()
```

```python
def clean_texts(text):
    original_text = text
    if pd.isna(text) or not str(text).strip() or str(text).strip().lower() == 'aucun':
        return "aucun"

    text_str = str(text).strip()


    if text_str.startswith("[") and text_str.endswith("]"):
        try:
            parsed_list = ast.literal_eval(text_str)
            if isinstance(parsed_list, list) and all(isinstance(item, str) for item in parsed_list):

                return ", ".join(item.strip() for item in parsed_list).lower()
            else:
                pass
        except (ValueError, SyntaxError) as e:
            pass

    tags = re.findall(r"['\"]([^'\"]*)['\"]", text_str)
    if tags:
        return ", ".join(tag.strip() for tag in tags).lower()
    return text_str.lower()

df['highlights'] = df['highlights'].apply(clean_texts)

# Display unique values to verify the transformation
df['highlights'].unique()
```

```python
df['ingredients']=df['ingredients'].apply(clean_texts)
df['ingredients'].head()
```

```python
def standardize_size(size):
    if pd.isna(size):
        return None
    size_str = str(size).lower()
    
    # Convertir toutes les unités en oz ou mL
    if 'oz' in size_str and 'ml' not in size_str:
        # Garder le format oz
        return re.sub(r'[^\d. oz/]', '', size_str).strip()
    elif 'ml' in size_str or 'mL' in size_str:
        # Extraire la valeur numérique
        match = re.search(r'(\d+(?:\.\d+)?)\s*(?:ml|mL)', size_str)
        if match:
            ml_value = float(match.group(1))
            # Convertir en oz si < 30ml pour standardisation
            if ml_value < 30:
                oz_value = ml_value / 29.5735
                return f"{oz_value:.1f} oz"
            else:
                return f"{ml_value} mL"
    return size_str

df['size'] = df['size'].apply(standardize_size)

# --- Créer des colonnes numériques pour la taille ---
def extract_numeric_size(size):
    if pd.isna(size):
        return None
    match = re.search(r'(\d+(?:\.\d+)?)', str(size))
    return float(match.group(1)) if match else None

df['size_numeric'] = df['size'].apply(extract_numeric_size)
df['size_unit'] = df['size'].apply(lambda x: 'oz' if x and 'oz' in str(x).lower() else ('mL' if x and 'ml' in str(x).lower() else 'unit'))
```

```python
df[['secondary_category', 'tertiary_category']] = df[['secondary_category', 'tertiary_category']].fillna('')
```

```python
df['size'].unique()
```

```python
# --- CRÉATION DE GROUPES DE NOTES ---

def categorize_rating(score):
    # Si le produit est "Top" (4.5 et plus)
    if score >= 4.5:
        return 'Top Rated'
    # Si le produit est "Bon" (entre 3.5 et 4.5)
    elif score >= 3.5:
        return 'Good'
    # Le reste (moins de 3.5, incluant les 0/Nouveaux)
    else:
        return 'Average'

print("Création de la colonne 'rating_group'...")
df['rating_group'] = df['rating'].apply(categorize_rating)
```

```python
columns_to_keep = [
    'product_id', 'product_name', 'brand_name',
    'price_usd', 'size',
    'rating', 'reviews', 'loves_count',
    'ingredients', 'highlights',
    'primary_category', 'secondary_category', 'tertiary_category',
    'new', 'out_of_stock','rating_group'
]

df_final = df[columns_to_keep]
```
